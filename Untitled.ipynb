```python
def remove_outlier(col):
    sorted(col)
    Q1,Q3=col.quantile([0.25,0.75])
    IQR = Q3-Q1
    lower_range = Q1-(1.5*IQR)
    upper_range = Q3+(1.5*IQR)
    return lower_range,upper_range
```

```python
lowprice,uppprice = remove_outlier(train['list_price'])
train['list_price']=np.where(train['list_price']>uppprice,uppprice,train['list_price'])
train['list_price']=np.where(train['list_price']<lowprice,lowprice,train['list_price'])
```
